```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('temperatures.csv')
df
```

```python
# input data
x=df['YEAR']
# output data
y=df['ANNUAL']
```

```python
#plt.figure(figsize=(16,9))
plt.title('Temperature plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual Avg temperature')
plt.scatter(x,y)
```

```python
x.shape
```

```python
#X as input variable , It must contain 2 values.
#convert x in array    "values" nam ka ak series attribute hai
x=x.values
```

```python
x.reshape(117,1)
```

```python
x=x.reshape(117,1)
```

```python
x.shape
```

```python
#Sci-kit learn officall website
#sci-kit leran  is the famous library in regression
from sklearn.linear_model import LinearRegression
```

```python
#create object name as regressor
regressor=LinearRegression() #constructor call
```

```python
#For algorithm train, we need fit(input,output)method for supervisied machine learning algorithm
regressor.fit(x,y)
```

```python
#Slop or coefficient parameter
#y = w.x + b
regressor.coef_ #w
```

```python
regressor.intercept_ #b
```

```python
#change the year n chk the tempt predivtion chng values
#Avg Temp
regressor.predict([[2078]])
```

```python
predicted=regressor.predict(x)
```

```python
#Using the Y=w.x+b (Coefficient and intercept value it will predict the following array)
predicted
```

```python
#Atcual value chk
y
```

```python
#MAE Mean absoulte error
y-predicted
```

```python
#if only need absolute value then use abs function
abs(y-predicted)
```

```python
np.mean(abs(y-predicted))
```

```python
# OR  without using formula we can do this with function 
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted)
```

```python
#MSE Mean square error
np.mean((y-predicted)**2)
```

```python
# OR  without using formula we can do this with function 
from sklearn.metrics import mean_squared_error
mean_squared_error(y,predicted)
```

```python
# R-Square error
from sklearn.metrics import r2_score
r2_score(y,predicted)
```

```python
# OR Using Score method
regressor.score(x,y)
```

```python
#plt.figure(figsize=(16,9))
plt.title('Temperature plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual Avg temperature')
plt.scatter(x,y,label='actual',color='r',marker='.')
plt.plot(x,predicted,label='predicted',color='g')# Line of best fit(Green line in below graph using regression equation Y=w.x+b)
plt.legend()
```
